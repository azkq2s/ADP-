### 6. 데이터프레임 값 변경 및 추가

```python
#선택한 값 변경 # 주로 이상치 값 변경
import pandas as pd

# 임의의 데이터프레임 생성
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 40, 45],
    'City': ['New York', 'Paris', 'London', 'Tokyo', 'Sydney'],
    'Score': [85, 92, 78, 88, 95]
}
my_df = pd.DataFrame(data)
my_df.loc[3, 'Score'] = 30
my_df['Age'] = my_df['Age']  + 10
my_df
```

```python
import pandas as pd

# 임의의 데이터프레임 생성
my_df = pd.DataFrame({
    'Name': ['Alice', 'Bob'],
    'Age': [25, 30],
    'City': ['New York', 'Paris']
})

new_row = pd.Series({
    'Name': 'Charlie',
    'Age': 35,
    'City': 'London'
})

# pd.concat()을 사용하여 행 추가
# ignore_index=True를 사용하여 인덱스를 재설정합니다.
# pd.concat()은 원본 DataFrame을 변경하지 않고 새로운 DataFrame을 반환합니다.
my_df_appended = pd.concat([my_df, new_row.to_frame().T], ignore_index=True)

print("DataFrame after appending new row (using pd.concat):")
display(my_df_appended)

#append 사용시      df.append는 pandas 1.4.0버전부터 사용 x
#my_df = mt_df.append(new_row)
```

```python
#col 추가
new_column_values = ['Male', 'Female']

# my_df에 'Gender'라는 이름의 새 컬럼 추가
my_df['Gender'] = new_column_values

print("Gender 컬럼이 추가된 my_df:")
display(my_df)
```

```python
#삭제
#DF.drop(index=None, columns = None, inplace = False)
my_df_appended.drop(0, inplace=True)
my_df_appended.drop(columns= 'City',inplace=True)
my_df_appended
#inplace=True 잘 안 씀
```
