## 데이터 전처리 예제

이 노트북에서는 데이터 전처리의 다양한 기법들을 예시를 통해 살펴보겠습니다.

### 예제 데이터 생성

`scikit-learn`의 `make_classification`을 사용하여 분류 모델 학습에 사용할 가상의 데이터셋을 생성합니다.

```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_classification

# 가상의 데이터셋 생성
X, y = make_classification(
    n_samples=1000,          # 총 샘플 수
    n_features=10,           # 독립 변수의 수
    n_informative=5,         # 유용한(정보가 있는) 독립 변수의 수
    n_redundant=2,           # 다른 변수들과 선형 결합 관계에 있는 독립 변수의 수
    n_repeated=0,            # 반복되는 독립 변수의 수
    n_classes=2,             # 타겟 클래스의 수
    n_clusters_per_class=1,  # 각 클래스당 클러스터 수
    weights=[0.9, 0.1],      # 클래스 불균형을 위한 가중치 (클래스 0: 90%, 클래스 1: 10%)
    flip_y=0.01,             # 클래스 라벨이 잘못될 확률
    random_state=42          # 재현 가능성을 위한 시드
)

df = pd.DataFrame(X, columns=[f'feature_{i}' for i in range(X.shape[1])])
df['target'] = y

print("생성된 데이터셋의 형태:", df.shape)
print("타겟 변수의 클래스 분포:")
print(df['target'].value_counts())
display(df.head())
```

### 1. 표본 추출 (Sampling)

전체 데이터셋에서 특정 기준에 따라 일부 데이터를 선택하는 과정입니다. 데이터의 특성을 잘 반영하면서 효율적인 분석을 위해 사용됩니다.

#### 1.1 단순 랜덤 추출 (Simple Random Sampling)

데이터셋에서 각 개체가 동일한 확률로 선택되는 방법입니다. `sample()` 함수를 사용하여 구현할 수 있습니다.

```python
print("원본 데이터셋 크기:", len(df))

# 100개의 샘플을 단순 랜덤 추출
srs_sample = df.sample(n=100, random_state=42)

print("단순 랜덤 추출 샘플 크기:", len(srs_sample))
display(srs_sample.head())
```

#### 2.2 계통 추출 (Systematic Sampling)

데이터를 정렬한 후, 일정한 간격(k)으로 샘플을 추출하는 방법입니다. 시작점은 랜덤하게 선택할 수 있습니다.

```python
# 데이터를 인덱스 기준으로 정렬 (이미 정렬되어 있다고 가정)
# df_sorted = df.sort_values(by='feature_0').reset_index(drop=True)

k = 10 # 10개마다 하나씩 추출

# 시작점 랜덤 선택
start_index = np.random.randint(0, k)

systematic_sample = df.iloc[start_index::k]

print("계통 추출 샘플 크기:", len(systematic_sample))
display(systematic_sample.head())
```

#### 2.3 집락 추출 (Cluster Sampling)

전체 모집단을 여러 개의 집락(cluster)으로 나누고, 이 중 일부 집락을 무작위로 선택하여 해당 집락에 속한 모든 개체를 조사하는 방법입니다.

예제에서는 임의로 `feature_0` 값을 기준으로 몇 개의 '집락'을 정의하고 추출해봅니다.

```python
# 설명을 위해 임의로 집락을 정의합니다 (예: feature_0 값의 구간별로).
# 실제 데이터에서는 지역, 학교 등 자연스러운 집락이 존재합니다.
df['cluster_id'] = pd.cut(df['feature_0'], bins=5, labels=False)

# 고유한 집락 ID 확인
cluster_ids = df['cluster_id'].unique()
print("생성된 집락 ID:", cluster_ids)

# 2개의 집락을 무작위로 선택
selected_clusters = np.random.choice(cluster_ids, size=2, replace=False)
print("선택된 집락 ID:", selected_clusters)

cluster_sample = df[df['cluster_id'].isin(selected_clusters)]

print("집락 추출 샘플 크기:", len(cluster_sample))
display(cluster_sample.head())
```

#### 2.4 층화 추출 (Stratified Sampling)

모집단을 몇 개의 동질적인 층(stratum)으로 나누고, 각 층에서 필요한 수만큼 단순 랜덤 추출하는 방법입니다. 주로 타겟 변수의 분포를 원본과 유사하게 유지할 때 사용됩니다.

```python
from sklearn.model_selection import train_test_split

# 타겟 변수(target)를 층으로 사용
# 각 층(클래스)에서 비율을 유지하며 샘플 추출

# 전체의 10%를 층화 추출 (train_test_split을 활용하여 비율 추출)
# stratify=df['target'] 옵션으로 타겟 변수의 클래스 비율을 유지합니다.
stratified_sample, _ = train_test_split(
    df, test_size=0.9, stratify=df['target'], random_state=42
)

print("층화 추출 샘플 크기:", len(stratified_sample))
print("층화 추출 샘플의 타겟 변수 분포:")
print(stratified_sample['target'].value_counts(normalize=True))
print("원본 데이터셋의 타겟 변수 분포:")
print(df['target'].value_counts(normalize=True))
display(stratified_sample.head())
```

### 2. 데이터 분할 (Data Splitting)

모델 학습 및 평가를 위해 데이터를 여러 세트로 나누는 과정입니다. 과적합 방지 및 모델 성능을 객관적으로 평가하는 데 필수적입니다.

#### 3.1 `train_test_split` (홀드아웃 방법)

가장 기본적인 방법으로, 데이터를 훈련 세트와 테스트 세트로 한 번만 나눕니다. `test_size`나 `train_size`로 비율을 지정할 수 있습니다. `stratify` 옵션을 사용하여 타겟 변수의 클래스 비율을 유지할 수 있습니다.

*oversampling은 split 이후에 train에 적용시켜서 모델의 파라미터를 민감하게 조정시킴. 즉, 학습을 잘 시키기 위해서 train에만 오버샘플링.
그래서 `stratify`는 그냥 split할 때 무조건 해야함


```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print("훈련 세트 X 형태:", X_train.shape)
print("테스트 세트 X 형태:", X_test.shape)
print("훈련 세트 y 형태:", y_train.shape)
print("테스트 세트 y 형태:", y_test.shape)

print("\n훈련 세트 타겟 분포:\n", pd.Series(y_train).value_counts(normalize=True))
print("테스트 세트 타겟 분포:\n", pd.Series(y_test).value_counts(normalize=True))
```

#### 3.2 `ShuffleSplit` (무작위 순열 교차 검증)

데이터를 무작위로 섞은 후, 지정된 수만큼 훈련/테스트 세트를 반복적으로 생성합니다. 각 분할은 독립적이며, 데이터의 일부는 한 번도 선택되지 않을 수 있습니다 (`train_test_split`의 반복 실행과 유사하지만, `n_splits`만큼 분할 객체를 생성합니다).

```python
from sklearn.model_selection import ShuffleSplit

# 전체 데이터의 70%를 훈련, 30%를 테스트 세트로 5번 반복하여 분할
ss = ShuffleSplit(n_splits=5, test_size=0.3, random_state=42)

print("ShuffleSplit 분할:")
for i, (train_index, test_index) in enumerate(ss.split(X)):
    print(f"  분할 {i+1}: 훈련 샘플 수 = {len(train_index)}, 테스트 샘플 수 = {len(test_index)}")
```

#### 3.3 `KFold` (K-Fold 분할)

데이터를 K개의 동일한 크기의 '폴드(fold)'로 나눈 후, K번 반복하여 K-1개의 폴드를 훈련 세트로, 나머지 1개의 폴드를 테스트 세트로 사용합니다. 모든 데이터가 정확히 한 번씩 테스트 세트로 사용됩니다.

```python
from sklearn.model_selection import KFold

# 5-폴드 교차 검증
kf = KFold(n_splits=5, shuffle=True, random_state=42)

print("KFold 분할:")
for i, (train_index, test_index) in enumerate(kf.split(X)):
    print(f"  분할 {i+1}: 훈련 샘플 수 = {len(train_index)}, 테스트 샘플 수 = {len(test_index)}")
```

#### 3.4 `StratifiedKFold` (계층별 K-Fold 분할)

`KFold`와 유사하지만, 각 폴드에 타겟 변수의 클래스 비율이 원본 데이터셋과 동일하게 유지되도록 분할합니다. 클래스 불균형이 있는 데이터셋에 특히 유용합니다.

```python
from sklearn.model_selection import StratifiedKFold

# 5-폴드 계층별 교차 검증
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

print("StratifiedKFold 분할:")
for i, (train_index, test_index) in enumerate(skf.split(X, y)):
    y_train_fold, y_test_fold = y[train_index], y[test_index]
    print(f"  분할 {i+1}: 훈련 샘플 수 = {len(train_index)}, 테스트 샘플 수 = {len(test_index)}")
    print(f"    훈련 세트 타겟 분포: {pd.Series(y_train_fold).value_counts(normalize=True).to_dict()}")
    print(f"    테스트 세트 타겟 분포: {pd.Series(y_test_fold).value_counts(normalize=True).to_dict()}")
```

#### 3.5 `GroupKFold` (그룹별 K-Fold 분할)

데이터에 '그룹' 정보가 있을 때, 같은 그룹에 속하는 데이터가 훈련 세트와 테스트 세트에 동시에 포함되지 않도록 분할합니다. 예를 들어, 동일한 환자에게서 얻은 여러 샘플 데이터가 있을 경우 환자 ID를 그룹으로 사용할 수 있습니다.

예제에서는 `feature_1` 값을 기준으로 임의의 '그룹'을 생성하여 사용합니다.

```python
from sklearn.model_selection import GroupKFold

# 임의의 그룹 정보 생성 (예: feature_1 값의 구간별로 그룹 할당)
groups = pd.cut(df['feature_1'], bins=10, labels=False)

# 5-폴드 그룹별 교차 검증
gkf = GroupKFold(n_splits=5)

print("GroupKFold 분할:")
for i, (train_index, test_index) in enumerate(gkf.split(X, y, groups=groups)):
    train_groups = set(groups[train_index])
    test_groups = set(groups[test_index])
    # 훈련 세트와 테스트 세트의 그룹이 겹치지 않음을 확인
    assert len(train_groups.intersection(test_groups)) == 0
    print(f"  분할 {i+1}: 훈련 샘플 수 = {len(train_index)}, 테스트 샘플 수 = {len(test_index)}")
    print(f"    훈련 세트 고유 그룹 수: {len(train_groups)}, 테스트 세트 고유 그룹 수: {len(test_groups)}")
```

### 3. 교차 검증 (Cross-Validation)

데이터 분할 전략을 사용하여 모델의 성능을 여러 번 평가하고 평균을 내어 보다 신뢰성 있는 성능 추정치를 얻는 방법입니다.

#### 4.1 `sklearn.model_selection.cross_validate`

여러 가지 평가 지표를 동시에 계산하며 교차 검증을 수행할 수 있습니다. 훈련 시간, 테스트 시간 등 추가 정보도 제공합니다.

```python
from sklearn.model_selection import cross_validate
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score, f1_score

# 로지스틱 회귀 모델
model = LogisticRegression(random_state=42, solver='liblinear')

# 평가 지표 정의
scoring = {
    'accuracy': make_scorer(accuracy_score),
    'precision': make_scorer(precision_score),
    'recall': make_scorer(recall_score),
    'f1_score': make_scorer(f1_score)
}

# 5-폴드 계층별 교차 검증 수행
cv_results = cross_validate(
    model, X, y, cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=42),
    scoring=scoring, return_train_score=True
)

print("cross_validate 결과:")
for key, value in cv_results.items():
    print(f"{key}: {value.mean():.4f} (+/- {value.std():.4f})")
```

#### 4.2 `sklearn.model_selection.GridSearchCV`

모델의 하이퍼파라미터 튜닝을 위해 사용됩니다. 지정된 하이퍼파라미터 조합들을 교차 검증을 통해 모두 탐색하여 최적의 조합을 찾아줍니다.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

# Support Vector Machine 모델
svm_model = SVC(random_state=42)

# 탐색할 하이퍼파라미터 그리드 정의
param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto']
}

# GridSearchCV 설정
# cv=StratifiedKFold를 사용하여 클래스 불균형을 고려한 교차 검증
grid_search = GridSearchCV(
    svm_model, param_grid, cv=StratifiedKFold(n_splits=3, shuffle=True, random_state=42),
    scoring='accuracy', n_jobs=-1, verbose=1
)

# 하이퍼파라미터 탐색 실행
grid_search.fit(X, y)

print("\nGridSearchCV 결과:")
print("최적 하이퍼파라미터:", grid_search.best_params_)
print("최고 교차 검증 정확도:", grid_search.best_score_)

# 최적의 모델
best_svm_model = grid_search.best_estimator_
print("\n최적 모델로 예측:")
# X_test, y_test는 train_test_split에서 분할된 테스트 세트 사용
if 'X_test' in locals() and 'y_test' in locals():
    test_accuracy = best_svm_model.score(X_test, y_test)
    print(f"테스트 세트 정확도: {test_accuracy:.4f}")
else:
    print("테스트 세트가 정의되지 않아 예측을 수행할 수 없습니다. train_test_split을 먼저 실행해주세요.")
```
